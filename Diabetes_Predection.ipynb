# **Spilitting the Data**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print(x_train)
```

# **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

# **Trainng the  SVM Model**

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0)
classifier.fit(x_train,y_train)
```

# **Predecting Test Result**

```python
y_pred=classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

# **Accuracy_Score**

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
print('Accuracy is :',accuracy_score(y_test,y_pred))
```

# **Predecting a new test result for single instance**

```python
input_data=(6,148,72,35,0,33.6,0.627,50)


#changing data to numpy array
input_data=np.asarray(input_data)

#Reshaping our array as we are predecting only 1 instance
input_data_reshaped=input_data.reshape(1,-1)


#Now predecting for an single instance
single_pred=classifier.predict(sc.transform(input_data_reshaped))
print(single_pred)

if(single_pred[0]==0):
  print('Person isn\'t  Diabetic ')
else:
  print('Person is Daibetic')
```
